```python
# Q1: How do we store ordered lists and keyed dictionaries, then pull basic facts?
nums = [12, 7, 19, 7, 5]
profile = {"name": "Ava", "city": "Delhi", "skills": ['python', 'sql']}
print('List length:', len(nums))
print('Unique items:', sorted(set(nums)))
print('First three nums:', nums[:3])
print('Name:', profile['name'])
print('Has sql?', 'sql' in profile['skills'])
```

Lists keep ordered values; dictionaries map keys to values for fast lookup. Slicing, set() and membership checks let you summarize and query small collections quickly.

```python
# Q2: How do we create a DataFrame from dictionaries and set up plotting libraries?
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set(style='whitegrid')
sales = {"product": ['A', 'B', 'C'], "units": [120, 85, 140], "price": [10.0, 12.5, 9.0]}
df_sales = pd.DataFrame(sales)
print(df_sales)
```

Pandas turns Python dicts into tabular DataFrames; each column has a name and type. Seaborn/Matplotlib imports and styling prepare us for clean plots directly from frames.

```python
# Q3: How do we load the Titanic dataset and quickly inspect structure and stats?
titanic = sns.load_dataset('titanic')
preview = titanic.head(8)
summary = titanic.describe(include='all').transpose().loc[:,'count':'freq']
print('Preview:
', preview)
print('
Counts/Freqs (mixed types):
', summary.head(5))
```

Head() gives a tangible feel for rows; describe(include='all') mixes numeric and categorical stats. Quick scans catch missingness, categories, and value ranges before plotting.

```python
# Q4: What is the survival breakdown across passenger classes (stacked countplot)?
plt.figure(figsize=(6,3))
sns.countplot(data=titanic, x='class', hue='survived')
plt.title('Survival by Ticket Class')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```

Countplots compare category frequencies; hue stacks survival status to show composition. Higher survival in upper classes is visible without computing percentages first.

```python
# Q5: How is passenger age distributed, and is it skewed (histogram + KDE)?
plt.figure(figsize=(6,3))
sns.histplot(data=titanic, x='age', bins=30, kde=True, color='slateblue')
plt.title('Age Distribution with KDE')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

Histogram shows shape and gaps; KDE overlays a smooth density to reveal central mass and tails. Skew or multimodality guides later binning or transformations.

```python
# Q6: How do fares vary by ticket class, and where are the outliers (boxplot)?
plt.figure(figsize=(6,3))
sns.boxplot(data=titanic, x='class', y='fare')
plt.title('Fare by Class')
plt.xlabel('Class')
plt.ylabel('Fare')
plt.show()
```

Boxplots summarize medians, quartiles, and outliers; 1st class shows higher spread and extreme fares. Useful to compare numeric distributions across categories quickly.

```python
# Q7: Is there a relationship between age and fare, conditioned on survival (scatter)?
plt.figure(figsize=(6,3))
sns.scatterplot(data=titanic, x='age', y='fare', hue='survived', alpha=0.7)
plt.title('Age vs Fare by Survival')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

Scatterplots reveal correlation patterns and clusters; hue separates survival outcomes. Look for dense bands and outliers that may influence models.

```python
# Q8: What are survival rates by sex and class, visualized as a heatmap?
rate = titanic.pivot_table(values='survived', index='sex', columns='class')
plt.figure(figsize=(6,3))
sns.heatmap(rate, annot=True, fmt='.2f', cmap='Blues')
plt.title('Survival Rate by Sex & Class')
plt.xlabel('Class')
plt.ylabel('Sex')
plt.show()
```

Pivot tables compute survival proportions; heatmaps make differences pop via color. Females and higher classes show stronger survival in one glance.

```python
-- Q10: Show customers with their total spend using LEFT JOIN to keep zero-order customers.
SELECT c.customer_id, c.first_name, c.last_name,
       COALESCE(SUM(o.amount), 0) AS total_spend
FROM Customers c
LEFT JOIN Orders o ON c.customer_id = o.customer_id
GROUP BY c.customer_id, c.first_name, c.last_name
ORDER BY total_spend DESC;
```

Left joins retain all customers; COALESCE turns NULL into zero for non-buyers. Grouping by customer aggregates spend before sorting by value.

```python
-- Q11: Find products with at least 2 orders and average amount > 500 (HAVING filter).
SELECT item, COUNT(*) AS order_count, AVG(amount) AS avg_amount
FROM Orders
GROUP BY item
HAVING COUNT(*) >= 2 AND AVG(amount) > 500
ORDER BY avg_amount DESC;
```

HAVING filters groups post-aggregation, unlike WHERE. Combining count and average thresholds surfaces reliably popular, high-value items.
