# Plotting Essentials (Live Recap)
Walk through quick plotting patterns with Matplotlib/Seaborn; each code cell is followed by a short explanation.

## When to Use Each Plot (Quick Guide)
1) Line plot: trends over time or ordered sequence; best for showing change and continuity.
2) Bar chart: compare discrete categories or aggregates; keep categories on x-axis for clarity.
3) Column/bar stacked: compare parts-to-whole across categories; good for composition over groups.
4) Scatter plot: relationship between two numeric variables; spot correlation and clusters.
5) Histogram: distribution of one numeric variable; bin choice shows shape and skew.
6) KDE/density plot: smooth view of distribution; compare multiple groups with hue.
7) Box plot: median, quartiles, and outliers; great for numeric vs category comparisons.
8) Heatmap: 2D matrix values (e.g., correlations or pivot tables); color encodes magnitude.
9) Pie chart: rarely; only when showing simple part-of-whole with very few categories.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
sns.set(style='whitegrid')
```

Each choice above fits a common question: change over time (line), compare categories (bar), relationships (scatter), distributions (hist/KDE/box), matrix patterns (heatmap), and only seldom use pie for simple parts-of-whole. Pick the plot that answers the audience’s question fastest.

Load core libraries and enable inline plotting; Seaborn styling keeps defaults clean.

```python
# 1) Bar chart: category comparison
categories = ['A', 'B', 'C', 'D']
values = [18, 25, 12, 30]
plt.figure(figsize=(5,3))
plt.bar(categories, values, color='steelblue')
plt.title('Sales by Category')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.show()
```

Bar chart highlights categorical differences; titles/labels make the comparison clear. Great for quick categorical comparisons without overwhelming detail.

```python
# 2) Scatter plot: relationship check
import numpy as np
np.random.seed(0)
x = np.linspace(0, 10, 40)
noise = np.random.normal(0, 4, size=40)
y = 2 * x + 5 + noise
plt.figure(figsize=(5,3))
plt.scatter(x, y, color='darkorange', alpha=0.8)
plt.title('X vs Y (linear-ish)')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

Scatter plot shows correlation pattern and spread; randomness shows variance around a trend. Use it to spot clusters, outliers, and direction of association.

# EDA with Titanic Dataset (C1)
Use Seaborn's built-in Titanic data for quick EDA plots and summary stats.

```python
# 3) Load Titanic dataset
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
# 3a) Preview: show first 10 rows\ntitanic.head(10)
```

Load a ready dataset; inspect columns like age, fare, sex, class, survived. Head(10) gives a tangible feel for the rows before plotting; always sanity-check data before visuals.

```python
# 4) Bar plot: survival counts by class
plt.figure(figsize=(5,3))
sns.countplot(data=titanic, x='class', hue='survived')
plt.title('Survival by Class')
plt.xlabel('Ticket Class')
plt.ylabel('Count')
plt.show()
```

Count plot compares survival across classes; hue shows stacked survival status. Helps explain class impact on outcomes in a glance.

```python
# 5) Histogram/KDE: age distribution
plt.figure(figsize=(5,3))
sns.histplot(data=titanic, x='age', bins=30, kde=True, color='slateblue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.show()
```

Histogram + KDE shows central tendency and skew; gaps highlight missing ages. Use it to judge distribution shape before choosing summary stats.

```python
# 6) Box plot: fare by class
plt.figure(figsize=(5,3))
sns.boxplot(data=titanic, x='class', y='fare')
plt.title('Fare by Class')
plt.xlabel('Class')
plt.ylabel('Fare')
plt.show()
```

Box plot shows median and spread; outliers reveal very high fares in 1st class. Ideal for comparing numeric distributions across categories.

```python
# 7) Scatter: age vs fare colored by survival
plt.figure(figsize=(5,3))
sns.scatterplot(data=titanic, x='age', y='fare', hue='survived', alpha=0.7)
plt.title('Age vs Fare by Survival')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

Scatter plot hints how fare and age relate to survival; color separates outcomes. Useful to discuss multivariate patterns with a single view.

```python
# 8) Grouped mean: average age by survival
mean_age = titanic.groupby('survived')['age'].mean().round(2)
mean_age
```

Simple groupby reveals average age differences between survivors and non-survivors. Quick numeric summaries guide which cohorts to investigate further.

```python
# 9) Pivoted bar: survival rate by sex and class
rate = titanic.pivot_table(values='survived', index='sex', columns='class')
rate
```

Pivot table shows survival proportions; compare across sex and ticket class. Great for categorical cross-tabs without writing verbose groupby code.

```python
# 10) Heatmap of survival rates
plt.figure(figsize=(5,3))
sns.heatmap(rate, annot=True, fmt='.2f', cmap='Blues')
plt.title('Survival Rate by Sex & Class')
plt.show()
```

Heatmap visually emphasizes higher survival for females and higher classes. Color gradients make differences pop faster than raw tables.
